# AMAZON SALES DATASET ANALYSIS USING PDBC

```python
import sqlite3 as sql
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('amazonSalesdataset.csv',encoding="ISO-8859-1",index_col=[0])
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns=df.columns.str.replace('.','')
```

```python
df.head()
```

```python
conn=sql.connect('database.db')
c=conn.cursor()
```

```python
df.to_sql('AmazonSalesTable',conn)
```

```python
# displaying the AMAZONSALES table

c.execute('''select * from amazonsalestable;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the category of the products

c.execute('''select product_name,category
             from amazonsalestable
             group by product_name
             limit 100;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the distinct caterogies present in the database

c.execute('''select DISTINCT category
             from amazonsalestable;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the category and count of a particular product 

c.execute('''select product_name,category,count(*)
             from amazonsalestable
             group by category
             order by count(*) desc;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
c.execute('''select avg(actual_price)
             from amazonsalestable;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the top 20 products based on the ratings

c.execute('''select product_name,rating
             from amazonsalestable
             group by product_name
             order by rating desc
             limit 20;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the product and no.of ratings given

c.execute('''select product_name,rating_count
             from amazonsalestable;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the no.of products with 0 rating_count

c.execute('''select count(product_name) as ProductsWithZeroRatingCount
             from amazonsalestable
             where rating_count=0;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the products with their discount percent

c.execute('''select product_name,discount_percentage
             from amazonsalestable
             group by product_name
             order by discount_percentage desc;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the products where there is no discount

c.execute('''select count(product_name) as ProductsWithZeroDiscount
             from amazonsalestable
             where discount_percentage="0%";''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the average discount_percentage 

c.execute('''select avg(discount_percentage) AVGDiscountPercent
             from amazonsalestable;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays products with the comapred avg value w.r.t average discount_percenatge

c.execute('''select product_name,discount_percentage,
             (CASE
                WHEN discount_percentage<47.691468 THEN 'Below Average'
                WHEN discount_percentage=47.691468 THEN 'Average'
                WHEN discount_percentage>47.691468 THEN 'Above Average'
              END)as ComparedToAvgValue
             from amazonsalestable
             limit 5;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the products with discount percent > 90 

c.execute('''select product_name,actual_price,discounted_price,discount_percentage
             from amazonsalestable
             where discount_percentage>=90
             group by product_name
             order by discount_percentage desc;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the products with price>=500

c.execute('''select product_name,actual_price
             from amazonsalestable
             where actual_price>=" 500";''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the no.of boat products

c.execute('''select count(product_name) as BOAT_products
             from amazonsalestable
             where product_name like "%boat%";''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the customers with multiple purchases

c.execute('''select user_name,count(product_name) as No_of_purchases
             from amazonsalestable
             group by user_name
             order by No_of_purchases desc;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
#query displays the unique user id's

c.execute('''select distinct user_id
             from amazonsalestable;''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the products whose review is good

c.execute('''select product_name
             from amazonsalestable
             where review_title like "good%" or review_title like "%good%" or review_title like "%good";''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

```python
# query displays the products whose review is satisfied

c.execute('''select product_name
             from amazonsalestable
             where review_title like "satisfied%" or review_title like "%satisfied%" or review_title like "%satisfied";''')
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```
